```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_test = [(5, -3), (-3, 8), (3, 6), (0, 0), (5, 3), (-3, -1), (-3, 3)]
w=[-33,9,13]
predict=[]
x=[]
y=[]

#x1=[8,-5]
#y1=[-3,6]
x1=[i for i in range(-5,10)]
y1=[(33-9*i)/13 for i in x1]

for i in x_test:
    x.append(i[0])
    y.append(i[1])
    if w[0]+w[1]*i[0]+w[2]*i[1]>0:
        predict.append(1)
    else:
        predict.append(-1)
predict
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x1,y1)
plt.show()
```

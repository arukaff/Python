```python
from math import factorial, exp
def bernulli(n, k, p):
    comb=factorial(n)/(factorial(k)*factorial(n-k))
    return comb*(p**k)*(1-p)**(n-k)
def puasson(m,p,n):
    lamb=p*n
    return exp(-lamb)*(lamb**m)/factorial(m)
def combination(n,k):
    return factorial(n)//(factorial(k)* factorial(n-k))
```

1. Даны значения зарплат из выборки выпускников: 100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150. Посчитать (желательно без использования статистических методов наподобие std, var, mean) среднее арифметическое, среднее квадратичное отклонение, смещенную и несмещенную оценки дисперсий для данной выборки.

```python
z=[100, 80, 75, 77, 89, 33, 45, 25, 65, 17, 30, 24, 57, 55, 70, 75, 65, 84, 90, 150]
m=sum(z)/len(z)
print(f'Среднее арифметическое значения зарплат из выборки выпускников = {round(m,4)}')
b=sum([(i-m)**2 for i in z]) # считаем квадрат разнасти значения и среднего арихметического
print(f'Среднее квадратичное отклонение значения зарплат из выборки выпускников = {round((b/len(z))**(1/2),4)}')
print(f'Cмещенная оценка дисперсии значений зарплат из выборки выпускников = {round(b/len(z),4)}')
print(f'Несмещенная оценка дисперсии значений зарплат из выборки выпускников = {round(b/(len(z)-1),4)}')
```

2. В первом ящике находится 8 мячей, из которых 5 - белые. Во втором ящике - 12 мячей, из которых 5 белых. Из первого ящика вытаскивают случайным образом два мяча, из второго - 4. Какова вероятность того, что 3 мяча белые?

```python
P1=(combination(3,2)/combination(8,2))*(combination(5,3)*combination(7,1)/combination(12,4)) # вероятность, что из первой карзины не вытащили ни одного белого меча
P2=(combination(5,1)*combination(3,1)*combination(5,2)*combination(7,2))/(combination(8,2)*combination(12,4)) # вероятность, что из первой карзины вытащили один белый мяч
P3=(combination(5,2)*combination(5,1)*combination(7,3))/(combination(8,2)*combination(12,4))  # вероятность, что из первой карзины вытащили два белых мяча
print(f'Вероятность того, что из вытащенных мячей 3 белые  = {P1+P2+P3: .4f}')
```

3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: 

a). первым спортсменом 

б). вторым спортсменом 

в). третьим спортсменом

```python
PB=1/3 # мишень поражена одним выстрелом из трех
PA=PB*0.9+PB*0.8+PB*0.6 # полная вероятность наструпления события - попал первый или второй или третий
print(f'Вероятность того, что выстрел произвёл первый спортсмен: {PB*0.9/PA: .4f};\n'
      f'Вероятность того, что выстрел произвёл второй спортсмен: {PB*0.8/PA: .4f};\n'
      f'Вероятность того, что выстрел произвёл третий спортсмен: {PB*0.6/PA: .4f}.'
     )
```

4. Рост взрослого населения города X имеет нормальное распределение.
Причем, средний рост равен 174 см, а среднее квадратичное отклонение равно 8 см.

Какова вероятность того, что случайным образом выбранный взрослый человек имеет рост:

а). больше 182 см

б). больше 190 см

в). от 166 см до 190 см

г). от 166 см до 182 см

д). от 158 см до 190 см

е). не выше 150 см или не ниже 190 см

ё). не выше 150 см или не ниже 198 см

ж). ниже 166 см.

```python
def z(h):
    return (h-174)/8
print(f'а) Вероятность P(X>=182) = 1- табличное значение = {1- 0.8413: .4f} от Z = {z(182): .4f}')
print(f'б) Вероятность P(X>=190) = 1- табличное значение = {1- 0.9772: .4f} от Z = {z(190): .4f}')
print(f'в) Вероятность P(166<=X<=190) = табличное значение Z2 ({z(190): .4f}) - табличное значение Z1 ({z(166): .4f}) = {0.9772-0.1587: .4f} ')
print(f'г) Вероятность P(166<=X<=182) = табличное значение Z2 ({z(182): .4f}) - табличное значение Z1 ({z(166): .4f}) = {0.8413-0.1587: .4f} ')
print(f'д) Вероятность P(158<=X<=190) = табличное значение Z2 ({z(190): .4f}) - табличное значение Z1 ({z(158): .4f}) = {0.9772-0.0227: .4f} ')
print(f'е) Вероятность P(150<=X или x>=190) = 1- табличное значение Z2 ({z(190): .4f}) + табличное значение Z1 ({z(150): .4f}) = {1-0.9772+0.0013: .4f} ')
print(f'ё) Вероятность P(150<=X или x>=198) = 1- табличное значение Z2 ({z(198): .4f}) + табличное значение Z1 ({z(150): .4f}) = {1-0.9987+0.0013: .4f} ')
print(f'а) Вероятность P(X<=166) = табличное значение = {0.1587: .4f} от Z = {z(166): .4f}')
```

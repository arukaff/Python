Постройте графики для приведенных наборов данных. Найдите коэффициенты для линии
регрессии и коэффициенты детерминации. Что вы замечаете? Нанесите на график модель
линейной регрессии.


```python
X1= np.array([30,30,40, 40])
Y1= np.array([37, 47, 50, 60])
X2= np.array([30,30,40, 40, 20, 20, 50, 50])
Y2= np.array([37, 47, 50, 60, 25, 35, 62, 72])
X3 = np.array([30,30,40, 40, 20, 20, 50, 50, 10, 10, 60, 60])
Y3 = np.array([37, 47, 50, 60, 25, 35, 62, 72, 13, 23, 74, 84]) 

```

```python
n=len(X1)
b1=(n*np.sum(X1*Y1)-np.sum(X1)*np.sum(Y1))/(n*np.sum(X1**2)-np.sum(X1)**2)
b0=np.mean(Y1)-b1*np.mean(X1)
y_pred=b0+b1*X1
y_pred
plt.scatter(X1, Y1)
plt.title('r= ' + str(np.corrcoef(X1, Y1)[1][0]))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X1,y_pred)
plt.show()
```

```python
n=len(X2)
b1=(n*np.sum(X2*Y2)-np.sum(X2)*np.sum(Y2))/(n*np.sum(X2**2)-np.sum(X2)**2)
b0=np.mean(Y2)-b1*np.mean(X2)
y_pred=b0+b1*X2
y_pred
plt.scatter(X2, Y2)
plt.title('r= ' + str(np.corrcoef(X2, Y2)[1][0]))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X2,y_pred)
plt.show()
```

```python
n=len(X3)
b1=(n*np.sum(X3*Y3)-np.sum(X3)*np.sum(Y3))/(n*np.sum(X3**2)-np.sum(X3)**2)
b0=np.mean(Y3)-b1*np.mean(X3)
y_pred=b0+b1*X3
y_pred
plt.scatter(X3, Y3)
plt.title('r= ' + str(np.corrcoef(X3, Y3)[1][0]))
plt.xlabel('x')
plt.ylabel('y')
plt.plot(X3,y_pred)
plt.show()
```

```python
x = np.array([10,8,13,9,11,14,6,4,12,7,5])
y = np.array([8.04,6.95,7.58,8.81,8.33,9.96,7.24,4.26,10.84,4.82,5.68])

from sklearn.linear_model import LinearRegression
model = LinearRegression()
x=x.reshape(len(x),1)
model.fit(x,y)
r_sq = model.score(x, y)
print('coefficient of determination:', r_sq)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
plt.scatter(x, y)
plt.plot(x, model.intercept_+model.coef_*x)
plt.show()
```

```python
y_hat=model.intercept_+model.coef_*x
y_hat=y_hat.reshape(1,-1)
res=y-y_hat
plt.scatter(range(len(x)), res)
plt.axhline(np.mean(res))
stats.shapiro(res)
```

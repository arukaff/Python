```python
df = pd.read_csv('./laptop_price.csv', encoding='latin1')
df.head()
```

```python
df['Cpu'].value_counts()
```

```python
#def cpu_to_comp(x):
#    return str(x).split()[0]

#df['Cpu_Company'] = df['Cpu'].apply(cpu_to_comp)
df['Cpu_Company'] = df['Cpu'].apply(lambda i: i.split()[0])
df.head()
```

```python
df['Memory'].value_counts()
```

```python
def mem_to_gb(x):
    mem=str(x).split()[0]
    if mem.endswith('GB'):
        return int(mem.replace('GB',''))
    return int(float(mem.replace('TB',''))*1024)

df['Memory_Amount'] = df['Memory'].apply(mem_to_gb)

#df['Memory_Amount'] = df['Memory'].apply(lambda i: ''.join(filter(str.isdigit,i)))

df.head()
```

```python
df['Memory_Amount'].value_counts()
```

```python
d='512GB SSD +  1.0TB Hybrid'
print(d.replace(' ','').split('B')[-1])
```

```python
def mem_to_type(x):
    return x.replace(' ','').split('B')[-1]
    

df['Memory_Type'] = df['Memory'].apply(mem_to_type)
df.head()
```

```python
df.drop(columns=['Memory', 'ScreenResolution'], inplace=True)
```

```python
avg=df.pivot_table(index='bedrooms',
               values='price',
                  aggfunc='mean',
                  fill_value=0).sort_values(by='price', ascending=True)
avg
```

```python
price=df.groupby('condition').agg({'price': ['min', 'mean', 'max']})
price
```

```python
view_waterfront=df.pivot_table(index='waterfront',
                               columns='view',
               values='id',
                  aggfunc='count',
                  fill_value=0)
view_waterfront
```

```python
import numpy as np
# Создаем матрицу коэффициентов системы уравнений
A = np.array([[1, 2], [3, 4]])
# Создаем вектор правых частей системы уравнений
b = np.array([3, 6])
# Решаем систему уравнений методом Гаусса
x = np.linalg.solve(A, b)
# Выводим р
print(x)
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import seaborn as sns
# Определяем случайную матрицу 3x3
A = np.random.rand(3, 3)
# Применяем сингулярное разложение
U, B, VT = np.linalg.svd(A)
print("Матрица A:")
plt.figure(1)
sns.heatmap(A, annot=True, square=True, fmt='.2f', cbar=False)
print(A)
print("Матрица U:")
plt.figure(2)
sns.heatmap(U, annot=True, square=True, fmt='.2f', cbar=False)
print(U)
print("Матрица B:")
print(B)
#sns.heatmap(B, annot=True, square=True, fmt='.2f', cbar=False)
print("Матрица VT:")
print(VT)
fig = plt.figure()
ax1 = fig.add_subplot(121)
# Bilinear interpolation - this will look blurry
ax1.imshow(A, interpolation='bilinear', cmap=cm.Greys_r)
ax2 = fig.add_subplot(122)
# 'nearest' interpolation - faithful but blocky
ax2.imshow(A, interpolation='nearest', cmap=cm.Greys_r)
plt.show()
```
